```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../data/raw/train.csv")
print("Load data successfully!")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df['GrLivArea_scaled'] = scaler.fit_transform(df[['GrLivArea']])
```

```python
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']

correlation = df['TotalSF'].corr(df['SalePrice'])
print(correlation)
```

```python
missing_series = df.isnull().sum()
print(missing_series[missing_series > 0].sort_values(ascending=False))
```

```python
missing_cols = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']

for col in missing_cols:
    df[col] = df[col].fillna(df[col].median())

print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
numeric_cols = ['GrLivArea', 'TotalSF', 'HouseAge', 'RemodAge', 'TotalBsmtSF', 'LotFrontage', 'GarageYrBlt', 'MasVnrArea']

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

new_features = ['TotalSF', 'HouseAge', 'RemodAge']
print(df[new_features].describe().loc[['mean', 'std']])
```
