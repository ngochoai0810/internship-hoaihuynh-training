```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../data/raw/train.csv")

#5 first rows
df.head()
```

```python
df['SalePrice_log'] = np.log1p(df['SalePrice'])
fig , axes = plt.subplots(1,2,figsize=(16,5))
sns.histplot(df['SalePrice'], kde =True, ax=axes[0], color='blue')
sns.histplot(df['SalePrice_log'], kde =True, ax=axes[1], color='green')
```

```python
corr_matrix = df.corr(numeric_only=True)
top_features = corr_matrix['SalePrice'].abs().sort_values(ascending=False).head(10)
print(top_features)
```

```python
sns.scatterplot(data=df, x = 'GrLivArea', y = 'SalePrice')
plt.show()
```

```python
sns.boxplot(data=df, x="Neighborhood", y="SalePrice")
plt.xticks(rotation=45)
plt.show()
```
