```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../data/raw/train.csv")
print("Load data successfully!")
```

```python
# Calculate the correlation matrix for numerical features
numerical_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numerical_cols].corr()

# 2. Filter the correlation matrix to find pairs of features with high correlation (e.g., > 0.8)
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_pairs = [(column, row) for column in upper_tri.columns for row in upper_tri.index if abs(upper_tri[column][row]) > 0.8]
print("Các cặp biến có nguy cơ đa cộng tuyến cao (> 0.8):", high_corr_pairs)

# 3. Drop one feature from each pair of highly correlated features
cols_to_drop = ['GarageArea', '1stFlrSF', 'TotRmsAbvGrd']
df = df.drop(columns=cols_to_drop, errors='ignore')
print(f"Kích thước dataset sau khi dropped features đa cộng tuyến: {df.shape}")
```
