```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../data/raw/train.csv")
print("Load data successfully!")
```

```python
print(f"Kích thước dataset trước khi xử lý outlier: {df.shape}")

# defining a function to remove outliers based on the IQR method
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice')
plt.title('GrLivArea vs SalePrice (Identifying Outliers)')
plt.show()
```

```python
outliers = df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 300000)]
print(f"Số lượng dòng outlier nghiêm trọng bị loại bỏ: {len(outliers)}")
```

```python
df = df.drop(outliers.index)
print(f"Kích thước dataset sau khi xử lý outlier: {df.shape}")
```
